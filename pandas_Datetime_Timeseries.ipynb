### Part 1: Load and Inspect Data

```python
import pandas as pd
```

```python
df = pd.read_csv("ETH_1h.csv")
```

### Part 2: Transform "Date" to standard format

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p") # check datetime for specific formats
```

```python
df["Date"].head()
```

### Part 3: Generate new features from "Date", including "DayOfWeek" and "DateDelta"

```python
df.loc[0, "Date"].day_name()
```

```python
df["DayOfWeek"] = df["Date"].dt.day_name()
```

```python
df.head()
```

```python
filt = (df["Date"] >= "2020")
df.loc[filt]
```

```python
filt = (df["Date"] >= pd.to_datetime("2019-01-01")) & (df["Date"] < pd.to_datetime("2020-01-01"))
df.loc[filt]
```

### Part 4: Set "Date" as index, and slice df using "Date"

```python
df.set_index("Date", inplace=True) # Set "Date" to be index, then can be sliced directly using "Date"
```

### Part 5: Aggregate functions based on "Day", "Week", etc.

```python
df["2020-02-01"]["High"].max()
```

```python
highs = df["High"].resample("D").max()
Volumes = df["Volume"].resample("D").sum()
```

```python
%matplotlib inline
highs.plot()
```

```python
Volumes.plot()
```

```python
df.resample("W").mean()
```

```python
df.resample("W").agg({"Close": "mean", "High": "max", "Low": "min", "Volume": "sum"})
```
